# Linear Regression Model - This is the main model for the features I finalized

Predicting trip duration using user type and start hour as features

```python
# Convert user_type to float
bike["Is_Subscriber"] = bike['user_type'].map({'Customer':0, 'Subscriber':1})
```

# Train Test Split

```python
# Linear model - Using Trip_Duration as the response and Is-Subscriber and Start_Hour as the features

# create X and y
feature_cols = ['start_hour', 'Is_Subscriber']
X = bike[feature_cols]
y = bike.trip_duration

# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
bike.head()
```

```python
# Fitting a linear regression model and print coefficients
linreg = LinearRegression()
linreg.fit(X, y)
linreg.coef_
```

```python
# 10 MSE scores by cross validation
scores = cross_val_score(linreg, X, y, cv=10, scoring='neg_mean_squared_error')
```

```python
np.mean(np.sqrt(-scores))
```

```python
treereg = DecisionTreeRegressor(max_depth=1, random_state=1)
scores = cross_val_score(treereg, X, y, cv=10, scoring='neg_mean_squared_error')
np.mean(np.sqrt(-scores))
```

```python
treereg.fit(X,y)
pd.DataFrame(list(zip(feature_cols,treereg.feature_importances_)))
```

# Random Forest Regressor

```python
# Import class, instantiate estimator, fit with all data
rfreg = RandomForestRegressor(n_estimators=200, max_features=2, oob_score=True, random_state=1)
rfreg.fit(X_train, y_train)

# Caluclate classification accuracy
print((rfreg.oob_score_))
print(len(feature_cols))
print(rfreg.feature_importances_)
```

```python
scores = cross_val_score(rfreg, X, y, cv=10, scoring='neg_mean_squared_error')
np.mean(np.sqrt(-scores))
```

# Ridge Regression Model

```python
rreg = Ridge(alpha=0.1, normalize=True) 
rreg.fit(X_train, y_train) 
rreg.coef_ 
preds = rreg.predict(X_test) 
print ('RMSE.(Ridge reg.) =', np.sqrt(metrics.mean_squared_error(y_test, preds)))

# Calculate accuracy score
preds = [int(y) for y in preds]
y_test = [int(y) for y in y_test]
print (metrics.accuracy_score(y_test, preds))
```

```python
alpha_range = 10.**np.arange(-2, 3) 
rregcv = RidgeCV(normalize = True, scoring = 'mean_squared_error', alphas = alpha_range) 
rregcv.fit(X_train, y_train) 
print (rregcv.alpha_) 
preds = rregcv.predict(X_test) 
print ('RMSE (Ridge CV reg.) =', np.sqrt(metrics.mean_squared_error(y_test, preds)))
```

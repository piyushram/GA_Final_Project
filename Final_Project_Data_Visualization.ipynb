```python
# Import packages and functions
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# allow plots to appear in the notebook
%matplotlib inline
```

```python
# Read data file
bike = pd.read_csv(r'C:\Users\piyus\Desktop\GA_Project\Data\2017_bikedata\bike.csv', parse_dates=['start_time'])
```

```python
bike.describe()
```

```python
bike.hist(alpha=0.8, bins=10, figsize=(15, 20))
```

```python
# visualize correlation matrix in Seaborn using a heatmap
sns.heatmap(bike.corr())
```

```python
# Boxplot of casual vs registered users grouped by gender
bike.boxplot(column='age', by='user_type')
```

```python
# Pandas scatter plot to compare trip duration and age
bike[bike.trip_duration<1500].plot(kind='scatter', x='trip_duration', y='age', alpha=0.7);
```

```python
# Seaborn scatter plot with regression line to compare trip duration and age
sns.lmplot(x='trip_duration', y='age', data=bike, ci=10, aspect=1.5);
plt.xlim(0, 1200);
plt.ylim(0, 140);
plt.xticks(np.arange(0, 1200, 100))
plt.yticks(np.arange(0, 140, 20))
```

```python
# Trip_Duration mean plot for every "hour"
bike.groupby('start_hour').trip_duration.mean().plot()
```

```python
# Hourly rental trend for "User_Type=Customer"
bike[bike.user_type == 'Customer'].groupby('start_hour').trip_duration.mean().plot.barh()
```

```python
# Hourly rental trend for "User_Type=Subscriber"
bike[bike.user_type == 'Subscriber'].groupby('start_hour').trip_duration.mean().plot.barh()
```
